```python
import math
import copy
from itertools import combinations
```

```python
def es_cuadrado(n):
    raiz = math.sqrt(n)
    return int(raiz + 0.5) ** 2 == n
```

```python
def simplificar(d):
    cambio = False
    for i in range (len(d)):
        di=d[i]**2
        for j in range (len(d)):
            if i!=j:
                posibleCuadrado = di+(d[j]**2)
                if es_cuadrado(posibleCuadrado):
                    d[i]=math.sqrt(posibleCuadrado)
                    d.pop(j)
                    cambio=true
    if cambio:
        d= simplificar(d)
    return d
```

```python
def descomponerEnCuadrados(n):
    c=n
    suma=0
    descomposicion = []
    while suma<n:
        c = math.floor(math.sqrt(c))
        nexto = suma + c**2
        if(nexto <= n):
            suma=nexto
            descomposicion.append(c)
        #print(f"n: {nexto}, c:{c}, suma: {suma}")
    #print("simplificando ", descomposicion)
    descomposicion = simplificateOnes(descomposicion)
    #simplificarCuadrado(descomposicion)
    return descomposicion

print(descomponerEnCuadrados(10434541))
```
